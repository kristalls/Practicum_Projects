## Модель прогнозирования оттока клиентов

```python
X = fitness.drop(columns='Churn', axis=1)
y = fitness['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=0,
                                                    stratify = y
                                                   )#ваш код здесь
```

```python
model_lr = LogisticRegression(max_iter = 700)

model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)

print(
    'Accuracy:{:.2f} Precision:{:.2f} Recall:{:.2f}'.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred)
        )
    )
```

```python
model_rf = RandomForestClassifier()

model_rf.fit(X_train, y_train)
y_pred = model_rf.predict(X_test)
print(
    'Accuracy:{:.2f} Precision:{:.2f} Recall:{:.2f} '.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred)
        )
    )


```

```python
sc = StandardScaler()
X_sc = sc.fit_transform(X)
linked = linkage(X_sc, method = 'ward')
```

```python
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Hierarchial clustering for fitness')
plt.show() 
```

```python
def show_clusters_on_plot(df, x_name, y_name, cluster_name):
    plt.figure(figsize=(5, 5))
    sns.scatterplot(
        df[x_name], df[y_name], hue=df[cluster_name], palette='Paired'
    )
    plt.title('{} vs {}'.format(x_name, y_name))
    plt.show()
```

```python
km = KMeans(n_clusters=5, random_state=0)
# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 2)
labels = km.fit_predict(X_sc)
fitness['clusters'] = labels
clusters = fitness.groupby('clusters').mean()
# сохраняем метки кластера в поле нашего датасета

print(clusters)
# выводим статистику по средним значениям наших признаков по кластеру
# отрисуем графики для попарных сочетаний признаков с разметкой по кластерам
col_pairs = list(itertools.combinations(fitness.drop('clusters', axis=1).columns, 2))
for pair in col_pairs:
    show_clusters_on_plot(fitness, pair[0], pair[1], 'clusters')
    
```
